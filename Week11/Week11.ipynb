```python
# NAME: gradebook.py
# IMPORT STATEMENTS
# None


# FUNCTION DEFINITIONS

# Function 1: add_students(gradebook_dict)
def add_students(gradebook_dict):
  # Goal: Use a while loop to get names and grades.
  while True:
    name = input("Enter student name (or 'done'): ")
    # End when user types "done".
    if name.lower() == "done":
      break
    try:
      grade = float(input(f"Enter {name}'s grade: "))
      # The student's name is the key, and their grade is the value.
      gradebook_dict[name] = grade
    except ValueError:
      print("Invalid input. Please enter a numerical score.")

# Function 2: calculate_average(gradebook_dict)
# Goal: Use .values() to get grades and return the average.
def calculate_average(gradebook_dict):
  if not gradebook_dict:
    return 0.0
  return sum(gradebook_dict.values()) / len(gradebook_dict)


# PROGRAM LOGIC
# 1. Initialize empty dictionary 'gradebook'
gradebook = {}

# 2. Call add_students to populate the dictionary.
add_students(gradebook)

# 3. Call calculate_average and print result
avg = calculate_average(gradebook)

# 4. Print the final report of all names and grades
print("\n--- Class Report ---")
for name,grade in gradebook.items():
  print(f"{name}: {grade:.0f}")
print(f"\nClass Average: {avg:.2f}")
```

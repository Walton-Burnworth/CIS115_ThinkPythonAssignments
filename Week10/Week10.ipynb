```python
# NAME: roster.py
# IMPORT STATEMENTS
# None


# FUNCTION DEFINITIONS

# Function 1: add_players
def add_players(player_list):
  # Parameters: list
  while True:
    name = input("Enter player name (or quit to stop): ").strip()
    # Logic: While loop to get names, append to list, stop if "quit"
    if name.lower() == "quit":
      break
    player_list.append(name)

# Function 2: on_roster
def on_roster(player_list, name_to_find):
  # Parameters: list, search_name
  if name_to_find in player_list:
    return True
  # Logic: Return True if search_name is in list, else False
  else:
    return False


# PROGRAM LOGIC
# 1. Initialize empty list 'roster'
roster = []

# 2. Call add_players(roster)
add_players(roster)

# 3. Prompt user for a search name
print("Current Roster List:", roster)
print()
search_query = input("Who would you like to search for? ").strip()

# 4. Call on_ roster and print result
if  on_roster(roster, search_query):
  print("Result: Player found!")
else:
  print("Resut: Player not found.")
```

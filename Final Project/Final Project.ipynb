```python
# NAME: expense_tracker.py

# IMPORT STATEMENTS
# None

# FUNCTION DEFINITIONS

# Function 1: calculate_total(expense_list, filter_type, filter_value)
# Goal: Loop through the list and return a filtered sum.
def calculate_total(expense_list, filter_type=None, filter_value=None):
    total = 0.0
    for expense in expense_list:
        # Check if we are filtering by Category
        if filter_type == "Category":
            if expense["category"].lower() == filter_value.lower():
                total += expense["amount"]
        # Check if we are filtering by Vendor
        elif filter_type == "Vendor":
            if expense["vendor"].lower() == filter_value.lower():
                total += expense["amount"]
        # No filter: Sum everything
        else:
            total += expense["amount"]
    return total

# PROGRAM LOGIC

# 1. Initialize empty list to hold expense dictionaries
master_expenses = []

# 2. Main Menu Loop
while True:
    print("\n--- Expense Tracker ---")
    print("1. Add New Expense")
    print("2. Get Totals")
    print("3. Exit")

    selection = input("Selection: ")

    if selection == "1":
        category = input("Enter Category: ")
        vendor = input("Enter Vendor: ")

        # Part 3 requirement: Program doesn't crash if user enters letters for amount
        try:
            amount = float(input("Enter Amount: "))
        except ValueError:
            print("Error: Please enter a numeric value for the amount.")
            continue

        date = input("Enter Date (e.g., 2026-04-01): ")

        expense_record = {
            "category": category,
            "vendor": vendor,
            "amount": amount,
            "date": date
        }
        master_expenses.append(expense_record)

    elif selection == "2":
        print("\n--- Totaling Options ---")
        print("A: All Expenses")
        print("C: Filter by Category")
        print("V: Filter by Vendor")
        choice = input("Choice: ").upper()

        if choice == "A":
            total = calculate_total(master_expenses)
            # Part 3 requirement: dollar amounts always show two decimal places
            print(f"The total for all expenses is: ${total:.2f}")

        elif choice == "C":
            cat_to_sum = input("Enter Category to sum: ")
            total = calculate_total(master_expenses, "Category", cat_to_sum)
            print(f"The total for {cat_to_sum} is: ${total:.2f}")

        elif choice == "V":
            ven_to_sum = input("Enter Vendor to sum: ")
            total = calculate_total(master_expenses, "Vendor", ven_to_sum)
            print(f"The total for {ven_to_sum} is: ${total:.2f}")

    elif selection == "3":
        print("Exiting Program...")
        break

# 3. Final Report (Optional summary)
print("\nFinal Session Summary:")
for item in master_expenses:
    print(f"{item['date']} | {item['vendor']} ({item['category']}): ${item['amount']:.2f}")
```

```python
# IMPORT STATEMENTS
# None


# FUNCTION DEFINITIONS
# Define calculate_rectangle(length, width)
def calculate_rectangle(length, width):
  # Calculate Area as length * width
  area = length * width
  # Calculate Perimeter as 2 * (length + width)
  perimeter = 2 * (length + width)
  # "Pack" them into a tuple by returning both variables: return area, perimeter
  return area, perimeter


# PROGRAM LOGIC
# Step 1: Ask the user to input a length and a width and convert to floats

# Step 2: Call the calculate_rectangle function using the user's input
# Unpack the returned tuple into my_area and my_perimeter

# Step 3: Print the results clearly for the user
print(calculate_rectangle(10, 5))
```

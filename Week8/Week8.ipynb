```python
# Import Statements
# None

# Function Definitions
# Define is_palindrome that accepts the word that is to be checked.
def is_palindrome(word):
  # Ensure that it checks the word regardless of capitalization.
  clean_word = word.lower()
  # Use string splicing to flip the word.
  reversed_word = clean_word[::-1]
  # Compare the original word to what the word is after it is reversed.
  if clean_word == reversed_word:
    return True
  else:
    return False
  # Return True if they match, False if they don't.


# Program Logic
# Ask the user for a word.
user_input = input("Enter a word to check: ")
# Run the function using that word.
if is_palindrome(user_input):
  print("This is a palindrome.")
else:
  print("Not a palindrome.")
# Check the result:
  # If True, print "This is a palindrome".
  # If False, print "Not a palindrome".
```

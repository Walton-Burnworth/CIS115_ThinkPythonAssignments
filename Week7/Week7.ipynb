```python
# Import Statements
# None

# Function Definitioins
# 1. Define verify_pin():
def verify_pin():
  # Create a loop that asks for a PIN until it equals "1234"
  pin = ""
  while pin != "1234":
    pin = input("Enter PIN: ")
  # Return True
  return True
# 2. Define withdraw(balance, amount):
def withdraw(balance, amount):
  # Check if enough money exists
  if amount <= balance:
    return balance - amount
  # Return the new (or old) balance
  else:
    print("Insufficient funds.")
    return balance

# Program Logic
# 1. Set current_balance to 500
current_balance = 500
# 2. Call verify_pin()
verify_pin()
# 3. Get withdraw amount from the users input.
amount_withdraw = int(input("Enter amount to withdraw: "))
# 4. Update current_balance by calling withdraw()
current_balance = withdraw(current_balance, amount_withdraw)
# 5. Print the final balance
print("Final balance:", current_balance)
```

```python
# Import Statements #
# Import math.pi in order to get an accurate value of pi.
import math


# Function Definitions #
# 1. Define area_funtion that takes radius and returns 2πr.
def area_function(radius):
  return math.pi * radius * radius

# 2. Define circumference_function that takes radius and returns 2πr.
def circumference_function(radius):
  return 2 * math.pi * radius

# 3. Define volume_function that takes radius and height.
def volume_function(radius, height):
  return area_function(radius) * height


# Program Logic #
# Get user input
# prompt the user for the radius and the height, and convert both to float.
radius = float (input("Enter the radius: "))
height = float (input("Enter the height: "))

# Call all three functions and print the results with labels.
print("Area:", area_function(radius))
print("Circumference:", circumference_function(radius))
print("Volume:", volume_function(radius, height))
```
